```python
# 导入必要的库
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # 导入 numpy 用于处理潜在的 Inf 值

# 设置 Matplotlib 支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

# --- 数据类型转换 ---

# 转换 '日期' 列为 datetime 对象
train_df['日期'] = pd.to_datetime(train_df['日期'])
test_df['日期'] = pd.to_datetime(test_df['日期'])

# 显示转换后的数据信息
print("\n训练数据信息 (日期转换后):")
train_df.info()
print("\n测试数据信息 (日期转换后):")
test_df.info()

# --- 调查异常值 ---
print("\n--- 异常值调查 (训练集) ---")

# 1. 调查负价格
negative_prices_train = train_df[(train_df['开盘'] < 0) | (train_df['收盘'] < 0) | (train_df['最高'] < 0) | (train_df['最低'] < 0)]
print(f"\n训练集中存在负价格的记录数: {len(negative_prices_train)}")
if not negative_prices_train.empty:
    print("训练集中包含负价格的示例记录:")
    print(negative_prices_train.head())

# 2. 调查极端 '涨跌幅' 和 '振幅'
extreme_change_train = train_df[(train_df['涨跌幅'] > 100) | (train_df['涨跌幅'] < -100) | (train_df['振幅'] < -100) | (train_df['振幅'] > 1000) ] # 振幅的阈值可以调整
print(f"\n训练集中存在极端 '涨跌幅'或'振幅' 的记录数: {len(extreme_change_train)}")
if not extreme_change_train.empty:
    print("训练集中包含极端 '涨跌幅'或'振幅' (按涨跌幅降序):")
    print(extreme_change_train.sort_values(by='涨跌幅', ascending=False).head())
    print("训练集中包含极端 '涨跌幅'或'振幅'  (按振幅降序):") # 之前振幅的min值非常小
    print(extreme_change_train.sort_values(by='振幅', ascending=True).head())

print("\n--- 进一步EDA (主要基于训练集 train_df) ---")

# 1. 唯一股票代码数量
unique_stocks_train = train_df['股票代码'].nunique()
print(f"\n训练集中唯一股票代码数量: {unique_stocks_train}")

# 2. 数据时间范围
min_date_train = train_df['日期'].min()
max_date_train = train_df['日期'].max()
print(f"训练集数据时间范围: 从 {min_date_train} 到 {max_date_train}")

min_date_test = test_df['日期'].min()
max_date_test = test_df['日期'].max()
print(f"测试集数据时间范围: 从 {min_date_test} 到 {max_date_test}")

# --- 处理或过滤极端值 ---
train_df_for_plotting = train_df[
    (train_df['开盘'] >= 0) &
    (train_df['收盘'] >= 0) &
    (train_df['最高'] >= 0) &
    (train_df['最低'] >= 0) &
    (train_df['涨跌幅'] > -100) & (train_df['涨跌幅'] < 100) &
    (train_df['振幅'] >= 0) & (train_df['振幅'] < 100)
].copy()

print(f"\n用于绘图的训练集数据条数 (过滤后): {len(train_df_for_plotting)}")
if len(train_df_for_plotting) < len(train_df):
    print(f"因过滤极端值或负价格，移除了 {len(train_df) - len(train_df_for_plotting)} 条记录用于绘图。")

# --- 3. 数值特征分布 ---
print("\n绘制数值特征分布图 (基于过滤后的训练数据)...")

numerical_cols = ['开盘', '收盘', '最高', '最低', '成交量', '成交额', '振幅', '涨跌额', '换手率', '涨跌幅']
for col in numerical_cols:
    if col in train_df_for_plotting.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(train_df_for_plotting[col], kde=True, bins=50)
        plt.title(f'{col} 分布图 (过滤后)')
        plt.xlabel(col)
        plt.ylabel('频数')
        plt.show()
        print(f"已显示 {col} 分布图")
    else:
        print(f"列 {col} 不在过滤后的数据中，跳过绘图。")

# --- 4. 特定股票收盘价时序图 ---
print("\n绘制特定股票收盘价时序图 (基于过滤后的训练数据)...")
sample_stock_codes = train_df_for_plotting['股票代码'].unique()[:3] # 取前3个不同股票代码作为样本
if len(sample_stock_codes) > 0:
    plt.figure(figsize=(14, 7))
    for stock_code in sample_stock_codes:
        stock_data = train_df_for_plotting[train_df_for_plotting['股票代码'] == stock_code]
        plt.plot(stock_data['日期'], stock_data['收盘'], label=f'股票代码: {stock_code}')
    plt.title('部分股票收盘价随时间变化图 (过滤后)')
    plt.xlabel('日期')
    plt.ylabel('收盘价')
    plt.legend()
    plt.grid(True)
    plt.show()
    print("已显示部分股票收盘价时序图")
else:
    print("过滤后的数据中没有足够的股票样本用于绘制时序图。")

# --- 5. 涨跌幅分布图 ---
if '涨跌幅' in train_df_for_plotting.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(train_df_for_plotting['涨跌幅'], kde=True, bins=50)
    # 限制x轴范围，因为过滤后仍然可能有一定的尾部
    plt.xlim(-20, 20)
    plt.title('涨跌幅 分布图 (过滤后)')
    plt.xlabel('涨跌幅 (%)')
    plt.ylabel('频数')
    plt.show()
    print("已显示涨跌幅分布图")

# --- 6. 特征相关性热力图 ---
print("\n计算并绘制特征相关性热力图 (基于过滤后的训练数据)...")
# 替换无穷大值为NaN，避免corr()计算错误
train_df_for_plotting.replace([np.inf, -np.inf], np.nan, inplace=True)

# 选择数值列进行相关性计算
correlation_matrix_cols = ['开盘', '收盘', '最高', '最低', '成交量', '成交额', '振幅', '涨跌额', '换手率', '涨跌幅']
# 确保所有列都存在于 train_df_for_plotting
existing_cols_for_corr = [col for col in correlation_matrix_cols if col in train_df_for_plotting.columns]

if len(existing_cols_for_corr) > 1 : # 需要至少两列才能计算相关性
    correlation_matrix = train_df_for_plotting[existing_cols_for_corr].corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('数值特征相关性热力图 (过滤后)')
    plt.show()
else:
    print("过滤后的数据中不足以计算相关性矩阵。")
```
